```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv("healthcare_data.csv")
```

```python
# Group and count occurrences of medical conditions by gender
condition_gender = data.groupby(['Medical Condition', 'Gender']).size().unstack()

# Plot a grouped bar chart
condition_gender.plot(kind='bar', figsize=(8, 5), colormap="viridis")
plt.title("Medical Conditions by Gender", fontsize=16)
plt.xlabel("Medical Condition", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Gender")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```

```python
# Plot age distribution
plt.figure(figsize=(8,5))
sns.histplot(data['Age'], bins=15, kde=True, color="skyblue")
plt.title("Age Distribution of Patients", fontsize=16)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Count occurrences of each blood type
blood_type_counts = data['Blood Type'].value_counts()

# Plot a pie chart
plt.figure(figsize=(8,5))
blood_type_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colormap='cool')
plt.title("Blood Type Distribution", fontsize=16)
plt.ylabel("")  # Remove default ylabel
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count the number of patients in each hospital
hospital_patient_counts = data['Hospital'].value_counts()

# Limit to top 10 hospitals (or adjust as needed)
top_hospitals = hospital_patient_counts.head(10)

# Plot a horizontal bar chart
plt.figure(figsize=(8, 5))  # Reduced figure size
top_hospitals.plot(kind='barh', color='skyblue', edgecolor='black')
plt.title("Top 10 Patient Distributions Across Hospitals", fontsize=14)
plt.xlabel("Number of Patients", fontsize=12)
plt.ylabel("Hospital", fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()

# Save the plot as an image
plt.savefig("hospital_patient_distribution.png", dpi=150)

# Optionally show the plot
plt.show()
top_hospitals

```

```python
# Filter relevant columns and drop missing values
filtered_data = data[['Age', 'Medical Condition']].dropna()

# Set the plot size
plt.figure(figsize=(8, 5))

# Create a violin plot to visualize the distribution of age for each medical condition
sns.violinplot(
    x='Medical Condition', 
    y='Age', 
    data=filtered_data, 
    palette='coolwarm', 
    scale='width', 
    inner='quartile'
)

# Add titles and labels
plt.title('Relationship Between Age and Medical Condition', fontsize=16)
plt.xlabel('Medical Condition', fontsize=12)
plt.ylabel('Age', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add grid lines for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Count the number of patients for each doctor
doctor_patient_counts = data['Doctor'].value_counts()

# Plot a horizontal bar chart for doctor workload
plt.figure(figsize=(10, 6))
doctor_patient_counts.plot(kind='barh', color='green', edgecolor='black')

# Add titles and labels
plt.title('Number of Patients per Doctor', fontsize=16)
plt.xlabel('Number of Patients', fontsize=12)
plt.ylabel('Doctor', fontsize=12)

# Add grid lines for better readability
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```
